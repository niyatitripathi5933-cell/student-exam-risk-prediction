```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/cleaned_student_data.csv")
df.head()
```

```python
sns.boxplot(x="Final_Result", y="Attendance", data=df)
plt.xlabel("Final Result (0=Fail, 1=Pass)")
plt.ylabel("Attendance %")
plt.title("Attendance vs Exam Result")
plt.show()
```

```python
sns.boxplot(x="Final_Result", y="Internal_Marks", data=df)
plt.title("Internal Marks vs Exam Result")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure()
sns.boxplot(x="Final_Result", y="Attendance", data=df)
plt.savefig("../reports/attendance_vs_result.png")
plt.close()
```
